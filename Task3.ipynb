```python
df = pd.read_csv("Dataset .csv")
```

```python
df.info()
```

```python
top_votes = df.sort_values(by="Votes", ascending=False).head(10)

plt.figure(figsize=(10,5))

sns.barplot(x="Votes", y="Restaurant Name", data=top_votes)

plt.title("Top 10 Restaurants by Votes")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Aggregate rating"], bins=10)

plt.title("Ratings Distribution")

plt.show()
```

```python
sns.countplot(x="Has Online delivery", data=df)

plt.title("Online Delivery Availability")

plt.show()
```

```python
sns.countplot(x="Price range", data=df)

plt.title("Price Range Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="Votes", y="Aggregate rating", data=df)

plt.title("Votes vs Ratings")

plt.show()
```

```python
correlation = df["Votes"].corr(df["Aggregate rating"])

print("Correlation:", correlation)
```
